```python
import pandas as pd
from google_play_scraper import Sort, reviews

# ID paket aplikasi WhatsApp
app_id = 'com.whatsapp'

# Batas jumlah ulasan yang ingin diambil
count = 10000

# Mengambil ulasan
result, continuation_token = reviews(
    app_id,
    lang='en',  
    country='us',
    sort=Sort.NEWEST, 
    count=count
)
```

```python
# Konversi ke DataFrame
df = pd.DataFrame(result)
```

```python
df
```

```python
# Memilih kolom yang diinginkan
df = df[['userName', 'score', 'content', 'at']]
```

```python
# Bersihkan baris kosong
df.dropna(subset=['content', 'score'], inplace=True)
```

```python
# Simpan ke dalam file CSV
df.to_csv('whatsapp_reviews.csv', index=False, encoding='utf-8-sig')

print(f"Berhasil menyimpan {len(df)} review ke dalam data whatsapp_reviews.csv")
```
